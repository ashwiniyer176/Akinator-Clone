## Data Cleaning and Preprocessing Notebook

```python
import numpy as np
import pandas as pd
```

```python
marvel_df=pd.read_csv("../akinator/data/marvel-wikia-data.csv")
dc_df=pd.read_csv("../akinator/data/dc-wikia-data.csv")
```

```python
marvel_df.head()
```

```python
dc_df.head()
```

```python
dc_df.columns
```

```python
drop_cols=['page_id']
dc_df.drop(columns=drop_cols,inplace=True)
marvel_df.drop(columns=drop_cols,inplace=True)
```

```python
column_names=['name', 'urlslug', 'ID', 'ALIGN', 'EYE', 'HAIR', 'SEX',
       'GSM', 'ALIVE', 'APPEARANCES', 'FIRST APPEARANCE', 'YEAR']
dc_df.columns=column_names
marvel_df.columns=column_names
```

```python
marvel_df['Company']="Marvel"
dc_df['Company']="DC"
```

```python
comics_df=pd.concat([marvel_df,dc_df],ignore_index=True)
```

```python
comics_df.isnull().sum()
```

```python
comics_df.to_csv("../akinator/data/comics-combined-data.csv",index=False)
```
